```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./tracking_data/MPIAB Argos white stork tracking (1991-2017).csv')
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
year_2010 = data[data['timestamp'].dt.year==2010]
unique_indiv = year_2010['individual-local-identifier'].unique()

import plotly.graph_objects as go

# Assuming `year_2010` is your DataFrame
# And `unique_indiv` contains the unique identifiers of individuals

# Initialize a figure with a mapbox base map
fig = go.Figure()

for indiv in unique_indiv:
    sub = year_2010[year_2010['individual-local-identifier'] == indiv]
    sub = sub.sort_values(by='timestamp')
    
    hover_text = sub['timestamp'].astype(str)

    # Add a trace for each individual
    fig.add_trace(go.Scattermapbox(
        lat=sub['location-lat'],
        lon=sub['location-long'],
        mode='lines+markers',  # This plots both the trajectory (line) and the points (markers)
        marker=dict(size=5),
        name=str(indiv),  # Individual identifier as the trace name
        text=hover_text,  # This will show the timestamp on hover
        hoverinfo='text'
    ))

# Set the layout of the map
fig.update_layout(
    mapbox=dict(
        style="carto-positron",  # or "open-street-map", "stamen-terrain", etc.
        zoom=3,  # Adjust zoom level according to your data
        center=dict(lat=year_2010['location-lat'].mean(), lon=year_2010['location-long'].mean())  # Centering the map
    ),
    margin={"r":0,"t":0,"l":0,"b":0}  # Remove margins for a full-screen map view
)



fig.show()
```
